## EDA of dataset1

### Examining missing values

```python
dataset1.isna().sum().to_frame()
```

#### Missing values analysis
- Based on missing values dataframe of dataset1
    - 1558 missing values in Pregnancy
        - handling considering 1 is female and male is 0, then fill pregancy with 0 for male 
        - incase female fill with -1 (unknown)
    - 242 missing values in alcohol_consumption_per_day
        - filling with 0 considering non-drinker
    - 92 missing values in Genetic_Pedigree_Coefficients
        - filling with median considering numeric variable
- Based on missing values dataframe of dataset2
    - There are missing values in Physical_activity
    - 3841 missing values in Physical_actvity

#### Handle missing values (pregnancy variable)

```python
dataset1.loc[(dataset1['Sex'] == 0) & (dataset1['Pregnancy'].isnull()), 'Pregnancy'] = 0
dataset1.loc[(dataset1['Sex'] == 1) & (dataset1['Pregnancy'].isnull()), 'Pregnancy'] = -1
dataset1.isna().sum().to_frame()
```

#### Handle missing values (alcohol_consumption_per_day)

```python
dataset1['alcohol_consumption_per_day'] = dataset1['alcohol_consumption_per_day'].fillna(0)
dataset1.isna().sum().to_frame()
```

#### Handle missing values (Genetic_Pedigree_Coefficient)

```python
median_GPC = dataset1['Genetic_Pedigree_Coefficient'].median()
dataset1['Genetic_Pedigree_Coefficient'] = dataset1['Genetic_Pedigree_Coefficient'].fillna(median_GPC)
dataset1.isna().sum().to_frame()
```
