```python
import pandas as pd
import json

# Your raw data
raw_data = {
    "Variable Label": {
        "0":"Patient Number","1":"Blood Pressure Abnormality","2":"Blood Pressure Abnormality",
        "3":"Level of Hemoglobin (g/dl)","4":"Genetic Pedigree Coefficient*","5":"Age",
        "7":"Sex","8":"Sex","15":"Level of Stress (Cortisol Secretion)"
    },
    "Value Labels": {
        "0":"Not Applicable","1":"0 = Normal","2":"1 = Abnormal",
        "3":"Not Applicable","4":"Not Applicable","5":"Not Applicable",
        "7":"0 = Male","8":"1 = Female","15":"1 = Low"
    }
}

df = pd.DataFrame(raw_data)

schema_dict = {}

for label, group in df.groupby('Variable Label'):
    mappings = {}
    is_continuous = False
    
    for val in group['Value Labels']:
        if "Not Applicable" in val:
            is_continuous = True
        elif "=" in val:
            parts = val.split("=")
            mappings[parts[0].strip()] = parts[1].strip()
    
    # Logic: If 'Not Applicable' was found, it's a numeric field
    # Otherwise, it's a categorical mapping
    if is_continuous:
        schema_dict[label] = "Continuous Numeric"
    else:
        schema_dict[label] = mappings

# Output the result
print(json.dumps(schema_dict, indent=4))
```
