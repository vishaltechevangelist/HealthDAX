## Creating llm config

```python
from pandasai_litellm.litellm import LiteLLM
llm = LiteLLM(model="ollama/llama3", api_base="http://localhost:11434")
```

## Importing smartdataframe

```python
from pandasai import SmartDataframe
sdf = SmartDataframe(df, config={"llm":llm, "verbose":True})
result = sdf.chat("Who are more chronic male or female")
```

```python
from pandasai import Agent

agent = Agent(df, config={"llm": llm})
agent.chat("Who can preganant male or female")

# Use this for Agent objects:
print(agent.last_code_executed)
```
